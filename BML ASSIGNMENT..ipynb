# DATA_PREPROCESSING


```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
```

```python
data=pd.read_csv("D:/BML/Stock_Market.csv")
data.head(16)
```

```python
data.isnull().sum()
```

```python
print(data.describe())
```

```python
print(data.info())
```

```python
data.fillna(4,inplace=True)
data.head(15)
```

```python
data.drop("Date",axis=1,inplace=True)
data.head(15)
```

```python
imputer=SimpleImputer(missing_values=np.nan,strategy='mean')
imputer=imputer.fit(data.iloc[:,1:])
data.iloc[:,1:]=imputer.transform(data.iloc[:,1:])
data.head(15)
```

```python
data['label']=data['Open'].shift(4)
data.head(10)
```

```python
data['label']=data['Open'].shift(4,fill_value=10)
data.head(10)
```

```python
data['High'] = data['High'].fillna(data['High'].mean())
data.head(11)
```

```python
scaler=StandardScaler()
```

```python
data=scaler.fit_transform(data.iloc[:,1:])
data
```
